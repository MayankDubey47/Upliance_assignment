```python
#merging the data sets
merged_data = pd.merge(df_user, df_session, on='User ID')
df = pd.merge(merged_data, df_order, on=['User ID', 'Session ID'])
```

```python
#rename some columns name
df.rename(columns={ 'Dish Name_x': 'Dish Name', 'Meal Type_x': 'Meal Type' }, inplace=True)
```

```python
# dropping unnecessary columns
df.drop(['Phone','Email','Dish Name_y','Meal Type_y'],axis=1,inplace=True)
```

```python
#Average order rating based on session rating
df.groupby('Dish Name')['Rating'].mean()
```

```python
#identify the favourite meal of the users
df['Favorite Meal'].value_counts()
```

```python
# indentify the location from most no of orders are placed
df.groupby('Location')['Total Orders'].sum().sort_values(ascending=False)
```

```python
# identify the user name who have placed the most order
df.groupby(['User ID','User Name'])['Total Orders'].sum().sort_values(ascending=False)
```

```python
# identify the most no of orders placed in which time of the day
df.groupby('Time of Day')['Total Orders'].sum().sort_values(ascending=False)
```

```python
# calculating average session rating per user
df.groupby(['User ID','User Name'])['Session Rating'].mean().sort_values(ascending=False)
```

```python
# Creating a bar plot to show the total orders by location
sns.barplot(df.groupby('Location')['Total Orders'].sum().head().sort_values(ascending=False),palette='crest')
plt.show()
```

```python
#Plotting a pie chart for user's favorite meals
plt.pie(df['Favorite Meal'].value_counts(),autopct="%1.1f%%",labels=df['Favorite Meal'].value_counts().index,explode=[0.01,0.01,0.01], colors = sns.color_palette('flare'))
plt.show()
```

```python
# Ploting bar chart to identify the most popular dish
plt.figure(figsize=(9,5))
sns.barplot(df['Dish Name'].value_counts().sort_values(ascending=False),palette='crest')
plt.title('Most Popular Dishes')
plt.ylabel('Order Count')
plt.xlabel('Dish Name')
plt.show()
```

```python
# Ploting location vs Amount spend
plt.figure(figsize=(9,5))
sns.barplot(df.groupby('Location')['Amount (USD)'].mean().sort_values(ascending=False),palette='crest')
plt.title('Average Spending by Location')
plt.xlabel('Location')
plt.ylabel('Average Spending (USD)')
plt.show()
```

```python
# Ploting bar chart for age groups vs amount spend
df['Age Group'] = pd.cut(df['Age'], bins=[18, 25, 35, 50], labels=['18-25', '26-35', '36-50'])
sns.barplot(df.groupby('Age Group')['Amount (USD)'].mean(),palette='crest')
```

```python
# Plotting the session duration vs. rating
plt.figure(figsize=(10, 6))
sns.lineplot(df.groupby('Duration (mins)')['Session Rating'].mean().reset_index(), x='Duration (mins)', y='Session Rating', marker='o')
plt.title('Session Duration vs Rating')
plt.xlabel('Duration (mins)')
plt.ylabel('Average Session Rating')
plt.show()
```

```python
# Creating a correlation matrix for numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns
correlation_matrix = df[numerical_cols].corr()

# Creating a heatmap to visualize the correlations
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
# Extract day of the week from Order Date
df['Order Day'] = df['Order Date'].dt.day_name()

order_day_trends = df.groupby('Order Day')['Order ID'].count().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=order_day_trends, x='Order Day', y='Order ID', order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Order Trends by Day of the Week')
plt.xlabel('Day')
plt.ylabel('Number of Orders')
plt.show()

```
